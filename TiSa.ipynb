```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
from scipy.stats import norm
import matplotlib.pyplot as plt

### read *.CSV data file
fLoc = "../Data/" 

fName0 = "tisaph.csv"
fName1 = "tisa_fluor.csv"
fName2 = "tisa_fluor_lasing.csv"
fName3 = "ndyag.csv"
fName4 = "hene.csv"
fName5 = "bulb.csv"

def plot(fName, xmin,xmax):
    data = np.genfromtxt(fLoc + fName , dtype=float , delimiter=',')
    x = data[:,0]
    y = data[:,1]

    plt.figure(figsize=(12, 8))
    plt.grid(color = "gray", linestyle="--")
    plt.legend( ) # add legend
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.xlim(xmin, xmax)
    plt.xlabel('Wavelength / nm', fontsize = 20) # x axis label
    plt.ylabel('Intensity', fontsize = 20) # y axis label

    plt.plot(x, y, color="red")
    plt.show()
```

```python
print("order of Saved time")
print("tisaph")
plot(fName0, 500, 1000)

print("tisa_fluor")
plot(fName1, 500, 1000)

print("tisa_fluor_lasing")
plot(fName2, 500, 1000)

print("ndyag")
plot(fName3, 500, 1000)

print("HeNe")
plot(fName4, 500, 1000)

print("bulb")
plot(fName5, 500, 1000)
```
