```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
import matplotlib.ticker as ticker
```

# Function

```python
def plot(path):

    RamanShift=[]
    Subtracted=[]

    with open(path) as f:
        d1 = [s.strip().split(";") for s in f.readlines()]
        d2 = d1[89::]
        for i in range(0,len(d2)):
            RamanShift.append(float(d2[i][3].replace(",",".")))
            Subtracted.append(float(d2[i][7].replace(",",".")))
            
    fig = plt.figure(figsize=(12, 8))
    plt.plot(RamanShift, Subtracted,"-", color="r")
    # plt.errorbar(angle, averagePower,  yerr = error, linestyle="None", capsize = 3)

    #x = np.linspace()
    #plt.xticks([0, 400, 800]) #目盛りの追加（データの番号）
    plt.xticks(fontsize = 20) #軸目盛変更
    plt.yticks(fontsize = 20) #軸目盛変更
    #plt.xticks(np.arange(0, 1001, 125), np.arange(0, 80, 10)) #最後が表示されない？
    #plt.yscale('log') #対数表示
    plt.grid(color = "gray", linestyle="--") #()の中なしでも可
    plt.minorticks_on() #補助目盛り
    plt.xlim(-300, 3000)
    plt.ylim(-300, 30000) #軸の範囲
    plt.xlabel("Raman shift (cm-1)",size = "xx-large", fontsize = 20)
    plt.ylabel("Photon count",size = "xx-large", fontsize = 20)
    # plt.legend()
    #plt.legend(loc="upper right") #ラベル位置指定
    plt.show()
    
    peakX = []
    dydx=[]
    for i in range(0, len(d2)-2):
        dydx.append(0)
        dydx[i] = (Subtracted[i+1]-Subtracted[i])/(RamanShift[i+1]-RamanShift[i])
        if i >10 and dydx[i-1] > 0 and dydx[i] < 0:
            if Subtracted[i] - np.average(Subtracted[i-10:i+10]) > 500:
                peakX.append(RamanShift[i])

#     for j in range(0,len(peak)-1):
#         if peak[j+1] - peak[j] > 20:
#             peak2.append(peak[j])
    print(peakX)
```
